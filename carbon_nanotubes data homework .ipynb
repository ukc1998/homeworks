```python
import os
import csv
import logging as lg
import mysql.connector as connection

# Removing the log file if it already exists
if os.path.exists("carbon_nanotubes_data.log"):
    os.remove("carbon_nanotubes_data.log")
    
lg.basicConfig(filename = "carbon_nanotubes_data.log", level = lg.INFO, format = "%(asctime)s - %(message)s")
console_log = lg.StreamHandler()
console_log.setLevel(lg.DEBUG)
format = lg.Formatter("%(asctime)s - %(message)s")
console_log.setFormatter(format)
lg.getLogger("").addHandler(console_log)
lg.info("Logging has been started!")

try:
    lg.info("Trying to establish the connection")
    conn = connection.connect(host = "localhost",
                              user = "root",
                             passwd = "Qwerty5@",
                             use_pure = True,
                             database = "ukc123"
                             )
    if conn.is_connected():
        lg.info("Connection established successfully with the database!")
        
        cursor = conn.cursor()
        query1 = "DROP TABLE IF EXISTS carbon_nanotubes"
        lg.info("Trying to drop the table if it already exists")
        cursor.execute(query1)
        
        lg.info("Trying to create the table")
        query2 = "CREATE table carbon_nanotubes(Chiral_indice_n INT(7), Chiral_indice_m INT(7), Initial_atomic_coordinate_u INT(7), Initial_atomic_coordinate_v INT(7), Initial_atomic_coordinate_w INT(7), Calculated_atomic_coordinates_u INT(7), Calculated_atomic_coordinates_v INT(7), Calculated_atomic_coordinates_w INT(7))"
        cursor.execute(query2)
        lg.info("Table created!")
        
            
        with open("carbon_nanotubes.csv", "r") as file:
            lg.info("Values insertion into table being started!")
            next(file)
            data = csv.reader(file, delimiter = "\n")
            
            for num, record in enumerate(data, start=1):
                rec = record[0].split(";")
                records_processed = []
                for i in rec:
                    if len(i.split(",")) > 1:
                        records_processed.append(int(i.split(",")[1]))
                    else:
                        records_processed.append(int(i))
                records_processed = ",". join([str(j) for j in records_processed])
                lg.info("Trying Pre_processing of data{i}".format(i = num))
                cursor.execute("INSERT INTO carbon_nanotubes VALUES({values})".format(values = records_processed))
                lg.info("Pre_processing of data{i} has been done".format(i = num))
        conn.commit()
        lg.info("All data has been successfully dumped!")
        
                    
except Exception as e:
    lg.error(str(e))
    conn.close()
    
finally:
    if conn.is_connected():
        cursor.close()
        conn.close()
        lg.info("Connection has been closed now!")
        lg.shutdown()
```
